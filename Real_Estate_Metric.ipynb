```python
# real_estate_metric.py
import pandas as pd

# Sample property data (replace with your real CSV if you have one)
data = {
    "Address": [
        "123 Main St, Sydney", "456 Elm St, Melbourne",
        "789 Oak St, Brisbane", "321 Pine St, Sydney"
    ],
    "Suburb": ["Sydney", "Melbourne", "Brisbane", "Sydney"],
    "Price_AUD": [1200000, 900000, 750000, 1350000],
    "Bedrooms": [3, 2, 3, 4],
    "Bathrooms": [2, 1, 2, 3],
    "Land_Size_sqm": [400, 250, 300, 450],
    "Weekly_Rent_AUD": [750, 500, 520, 800]
}

# Create DataFrame
df = pd.DataFrame(data)

# Calculate metrics
df["Price_per_sqm"] = df["Price_AUD"] / df["Land_Size_sqm"]
df["Rent_Yield_percent"] = (df["Weekly_Rent_AUD"] * 52 / df["Price_AUD"]) * 100

# Group by Suburb for investors' insight
suburb_summary = df.groupby("Suburb").agg(
    Avg_Price_AUD=("Price_AUD", "mean"),
    Avg_Price_per_sqm=("Price_per_sqm", "mean"),
    Avg_Rent_Yield_percent=("Rent_Yield_percent", "mean")
).reset_index()

# Show full property details and suburb summary
print("Individual Property Metrics:\n", df)
print("\nSuburb Summary Metrics:\n", suburb_summary)

# Optional: save results to CSV
df.to_csv("property_metrics.csv", index=False)
suburb_summary.to_csv("suburb_summary.csv", index=False)
```
